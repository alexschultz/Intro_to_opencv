```python
import os
from matplotlib import pyplot as plt
import cv2
import numpy as np
print('Using OpenCV version:', cv2.__version__)
```

# Working with Images

```python
# This is how you read an image into memory
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
```

```python
# get image shape
print('image shape {}'.format(lenaImage.shape))
```

```python
# displays an image in it's own window
cv2.imshow('lena', lenaImage)

# wait for any key to be pressed
cv2.waitKey(0)

# close all open windows
cv2.destroyAllWindows()
```

```python
# displaying an image like that isn't great for Jupyter Notebooks, lets use matplotlib instead
plt.axis("off")
plt.imshow(lenaImage)
```

```python
# Wondering why the color is all messed up? 
# Well it's because by default, OpenCV opens images using BGR 
# but Matplotlib is expoecting an RGB image. Lets fix that and try again!

plt.axis("off")
lenaRGBImage = cv2.cvtColor(lenaImage, cv2.COLOR_BGR2RGB)
plt.imshow(lenaRGBImage)
```

```python
# you can also convert to other color spaces using the cv2.cvtColor function. Let's turn Lena black and white
gray = cv2.cvtColor(lenaRGBImage, cv2.COLOR_RGB2GRAY)
plt.axis("off")
plt.imshow(gray)
```

```python
# This is getting riddiculous, but I wanted to point out another issue you might run into when trying to use Matplotlib to display images.
# To fix this, you can provide a cmap parameter to the plt 
plt.axis("off")
plt.imshow(gray, cmap='gray')
```

```python
# Let's look at some of the various filters we have in OpenCV
# https://docs.opencv.org/4.0.1/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1
# https://docs.opencv.org/4.0.1/d4/d13/tutorial_py_filtering.html

lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
# Remember to reverse the image channels for matplotlib
lenaRGBImage = cv2.cvtColor(lenaImage, cv2.COLOR_BGR2RGB)

# blur the image
blur = cv2.GaussianBlur(lenaRGBImage, (9, 9), 0)


f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(lenaRGBImage)
ax[1].imshow(blur)
```

```python
# Edge detection
# https://docs.opencv.org/4.0.1/da/d22/tutorial_py_canny.html
# the image must be in grayscale
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(lenaImage, 100, 200)
plt.axis("off")
plt.imshow(edges, cmap='gray')
```

```python

lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
lenaImage = cv2.cvtColor(lenaImage, cv2.COLOR_BGR2RGB)
b = lenaImage.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

g = lenaImage.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = lenaImage.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0


f, ax = plt.subplots(1, 3, sharey=True)
ax[0].imshow(r)
ax[1].imshow(g)
ax[2].imshow(b)
```

```python
# histograms
cv2.imshow('lena', lenaImage)
plt.hist(lenaImage.ravel(), 256, [0, 256])
plt.show()
```

```python
# show histogram for individual colors
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([lenaImage], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
plt.xlim([0, 256])
plt.show()
```

```python
# equalize the histogram to distribute pixel values accross range
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_GRAYSCALE)
lenaEqualized = cv2.equalizeHist(lenaImage)

plt.hist(lenaEqualized.ravel(), 256, [0, 256])
plt.hist(lenaImage.ravel(), 256, [0, 256])
plt.show()
```

```python
f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(lenaImage, cmap='gray')
ax[1].imshow(lenaEqualized, cmap='gray')
```

```python
# sometimes equalizing the histogram can have a detrimental effect on the image

img = cv2.imread(os.path.join('images', 'clahe.png'), 0)
equ = cv2.equalizeHist(img)
f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(img, cmap='gray')
ax[1].imshow(equ, cmap='gray')
```

```python
# in those cases we can use the localized histogram equalization
# CLAHE (Contrast Limited Adaptive Histogram Equalization)
img = cv2.imread(os.path.join('images', 'clahe.png'), 0)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl1 = clahe.apply(img)
f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(img, cmap='gray')
ax[1].imshow(cl1, cmap='gray')
```

```python
# selecting ROI (region of interest)
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
lenaImage = cv2.cvtColor(lenaImage, cv2.COLOR_BGR2RGB)
# [y1:y2, x1:x2]
roi = lenaImage[200:395, 206:365]
# resize image
largerRoi = cv2.resize(roi, None, fx=3, fy=3)

f, ax = plt.subplots(1, 3, sharey=True)
ax[0].imshow(lenaImage)
ax[1].imshow(roi)
ax[2].imshow(largerRoi)
```

```python
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
lenaImage = cv2.cvtColor(lenaImage, cv2.COLOR_BGR2RGB)

cv2.rectangle(lenaImage, (206, 200), (365, 395), (0, 255, 0), 1)
cv2.circle(lenaImage, (264, 271), 20, (255, 0, 0), 1)
cv2.circle(lenaImage, (337, 272), 20, (255, 0, 0), 1)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(lenaImage, 'Lena', (10, 500), font, 4, (0, 0, 255), 2, cv2.LINE_AA)

plt.axis("off")
plt.imshow(lenaImage)
```

```python
# image thresholding
# https://docs.opencv.org/4.0.1/d7/d4d/tutorial_py_thresholding.html
lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_GRAYSCALE)
lenaImage = cv2.medianBlur(lenaImage, 5)
lenaThreshold = cv2.adaptiveThreshold(lenaImage, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)


f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(lenaImage, cmap='gray')
ax[1].imshow(lenaThreshold, cmap='gray')
```

```python
# contours
# https://docs.opencv.org/4.0.1/d4/d73/tutorial_py_contours_begin.html

shapesImageOriginal = cv2.imread(os.path.join('images', 'shapes.png'))
shapesImageOriginal = cv2.cvtColor(shapesImageOriginal, cv2.COLOR_BGR2RGB)
shapesImage = shapesImageOriginal.copy()
gray = cv2.cvtColor(shapesImage, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray, 127, 255, 1)

contours, h = cv2.findContours(thresh, 1, 2)

for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01 * cv2.arcLength(cnt, True), True)
    print(len(approx))
    if len(approx) == 5:
        print("pentagon")
        cv2.drawContours(shapesImage, [cnt], 0, 255, 2)
    elif len(approx) == 3:
        print("triangle")
        cv2.drawContours(shapesImage, [cnt], 0, (0, 255, 0), 2)
    elif len(approx) == 4:
        print("square")
        cv2.drawContours(shapesImage, [cnt], 0, (0, 0, 255), 2)
    elif len(approx) == 9:
        print("half-circle")
        cv2.drawContours(shapesImage, [cnt], 0, (255, 255, 0), 2)
    elif len(approx) > 15:
        print("circle")
        cv2.drawContours(shapesImage, [cnt], 0, (0, 255, 255), 2)

f, ax = plt.subplots(1, 2, sharey=True)
ax[0].imshow(shapesImageOriginal)
ax[1].imshow(shapesImage)
```

```python
# using OpenCV High gui
# https://docs.opencv.org/4.0.1/da/d6a/tutorial_trackbar.html

lenaImage = cv2.imread(os.path.join('images', 'lena.png'), cv2.IMREAD_COLOR)
cv2.namedWindow('trackbar example')

val1 = 100
val2 = 100
def on_trackbar_val1_changed(val):
    val1 = val
    edges = cv2.Canny(lenaImage, val1, val2)
    cv2.imshow('trackbar example', edges)

def on_trackbar_val2_changed(val):
    val2 = val
    edges = cv2.Canny(lenaImage, val1, val2)
    cv2.imshow('trackbar example', edges)


cv2.createTrackbar('canny val1', 'trackbar example', 0, 200, on_trackbar_val1_changed)
cv2.createTrackbar('canny val2', 'trackbar example', 0, 200, on_trackbar_val2_changed)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
